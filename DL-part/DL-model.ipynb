```python

def extract_text_from_pdf(pdf_path):
    return extract_text(pdf_path)

def extract_skills(input_text):
    stop_words = set(nltk.corpus.stopwords.words('english'))
    word_tokens = nltk.tokenize.word_tokenize(input_text)

    filtered_skills_list = [str(skill).lower() for skill in skills_list if isinstance(skill, str)]
    print(filtered_skills_list)
 
    # remove the stop words
    filtered_tokens = [w for w in word_tokens if w not in stop_words]
 
    # remove the punctuation
    filtered_tokens = [w for w in word_tokens if w.isalpha()]

 
    # generate bigrams and trigrams (such as artificial intelligence)
    bigrams_trigrams = list(map(' '.join, nltk.everygrams(filtered_tokens, 2, 3)))
 
    # we create a set to keep the results in.
    found_skills = set()
 
    # we search for each token in our skills database
    for token in filtered_tokens:
        for i in filtered_skills_list:
            if token.lower() == i.lower():
                found_skills.add(token)
 
    # we search for each bigram and trigram in our skills database
    for ngram in bigrams_trigrams:
        for j in filtered_skills_list:
            if ngram.lower()== j.lower():
                found_skills.add(ngram)
 
    return found_skills
```

```python
import torch
from transformers import BertTokenizer, BertModel
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Initialize BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

def extract_skills_from_resume(resume_text):
    """
    Extract skills from the resume text using BERT embeddings.
    """
    # Tokenize and get BERT embeddings
    inputs = tokenizer(resume_text, return_tensors='pt', truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
    # Mean pooling to get sentence embeddings
    embeddings = outputs.last_hidden_state.mean(dim=1)
    return embeddings

def compute_similarity(embedding1, embedding2):
    """
    Compute similarity score between two embeddings.
    """
    embedding1 = embedding1.numpy()
    embedding2 = embedding2.numpy()
    similarity = cosine_similarity(embedding1, embedding2)
    return similarity[0][0]

def match_skills(resume_text, required_skills):
    """
    Match skills extracted from resume with required skills and compute similarity scores.
    Also compute the overall similarity score as an average.
    """
    # Extract skills from resume
    resume_embedding = extract_skills_from_resume(resume_text)

    # Compute similarity for each required skill
    similarity_scores = {}
    total_similarity = 0
    for skill in required_skills:
        skill_embedding = extract_skills_from_resume(skill)
        similarity = compute_similarity(resume_embedding, skill_embedding)
        similarity_scores[skill] = similarity
        total_similarity += similarity

    # Calculate overall similarity as a percentage
    overall_similarity = (total_similarity / len(required_skills)) * 100
    return similarity_scores, overall_similarity

# Example usage

required_skills = ["Python", "machine learning", "data analysis", "Java"]
resume_text=""
'''for i in skills:
    for j in required_skills:
        if  i == j:
            resume_text+=i+" "'''

resume_text="UI design"
print(resume_text)
similarity_scores, overall_similarity = match_skills(resume_text, required_skills)

print("Similarity Scores:")
for skill, score in similarity_scores.items():
    print(f"{skill}: {score:.4f}")

print(f"\nOverall Similarity Score: {overall_similarity:.2f}%")
```

```python
import spacy
from spacy.pipeline import EntityRuler
from transformers import BertTokenizer, BertModel
import torch
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load SpaCy model
nlp = spacy.load("en_core_web_sm")

# Define skills for the EntityRuler (you can extend this list)
skills = [
    "Python", "Machine Learning", "Data Analysis", "SQL", "React", "Django",
    "Project Management", "Git", "Docker", "TensorFlow", "Numpy", "Pandas"
]

# Add EntityRuler to recognize skills
patterns = [{"label": "SKILL", "pattern": skill} for skill in skills]
ruler = EntityRuler(nlp)
ruler.add_patterns(patterns)
nlp.add_pipe(ruler, before="ner")

# Sample resume text
resume_text = """
Experienced software developer skilled in Python, Machine Learning, and Data Analysis.
Proficient in SQL, React, and Django, with a strong background in managing projects using Git and Docker.
"""

# Process the resume text
doc = nlp(resume_text)

# Extract skills
extracted_skills = [ent.text for ent in doc.ents if ent.label_ == "SKILL"]
resume_skills_text = ', '.join(extracted_skills)
print("Extracted Skills from Resume:", resume_skills_text)

# Job descriptions to compare with (sample job descriptions)
job_descriptions = [
    "Looking for a Data Scientist with expertise in Python, Machine Learning, and SQL.",
    "Seeking a Web Developer proficient in React, Django, and Git.",
    "Hiring for a Project Manager with skills in Project Management, Docker, and Git."
]

# Load BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

def get_embedding(text):
    """Generate BERT embeddings for the given text."""
    # Tokenize the text with padding and truncation
    inputs = tokenizer(text, return_tensors='pt', padding='max_length', truncation=True, max_length=512)
    with torch.no_grad():  # Disable gradient calculation for inference
        outputs = model(**inputs)
    # Use mean pooling of the token embeddings to get a single vector representation
    embedding = outputs.last_hidden_state.mean(dim=1).squeeze()  # Squeeze to remove extra dimensions
    return embedding.numpy()

# Get embeddings for the resume and each job description
resume_embedding = get_embedding(resume_skills_text).reshape(1, -1)  # Reshape to 2D for cosine similarity
job_embeddings = [get_embedding(desc).reshape(1, -1) for desc in job_descriptions]

# Calculate cosine similarity between resume and each job description
similarities = [cosine_similarity(resume_embedding, job_emb)[0][0] for job_emb in job_embeddings]

# Sort job descriptions by similarity score
top_jobs = sorted(zip(job_descriptions, similarities), key=lambda x: x[1], reverse=True)[:3]

# Print the top 3 job descriptions with the highest similarity
print("\nTop 3 job descriptions with the highest similarity:")
for idx, (job, score) in enumerate(top_jobs, start=1):
    print(f"{idx}. Job Description: {job}\n   Similarity Score: {score:.4f}")
```

```python
import torch
import torch.nn as nn
from transformers import BertModel, BertTokenizer

class DualEncoderWithCrossAttention(nn.Module):
    def __init__(self, model_name='bert-base-uncased'):
        super(DualEncoderWithCrossAttention, self).__init__()
        self.resume_encoder = BertModel.from_pretrained(model_name)
        self.job_encoder = BertModel.from_pretrained(model_name)
        self.cross_attention = nn.MultiheadAttention(embed_dim=768, num_heads=8)
        self.fc = nn.Linear(768, 1)  # Final layer to output similarity score

    def forward(self, resume_input, job_input):
        # Encode resume and job descriptions separately
        resume_emb = self.resume_encoder(**resume_input).last_hidden_state  # (batch_size, seq_len, hidden_dim)
        job_emb = self.job_encoder(**job_input).last_hidden_state

        # Apply cross-attention between resume and job embeddings
        attn_output, _ = self.cross_attention(resume_emb, job_emb, job_emb)
        combined_emb = attn_output.mean(dim=1)  # Aggregate sequence into a single vector

        # Output similarity score
        similarity_score = self.fc(combined_emb)
        return similarity_score
```

```python
from torch.utils.data import DataLoader, Dataset
from transformers import BertTokenizer
import torch.optim as optim
import torch.nn.functional as F

class ResumeJobDataset(Dataset):
    def __init__(self, resumes, jobs, labels, tokenizer, max_length=128):
        self.resumes = resumes
        self.jobs = jobs
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        resume = self.resumes[idx]
        job = self.jobs[idx]
        label = self.labels[idx]

        resume_input = self.tokenizer(resume, padding='max_length', max_length=self.max_length, truncation=True, return_tensors="pt")
        job_input = self.tokenizer(job, padding='max_length', max_length=self.max_length, truncation=True, return_tensors="pt")

        # Squeeze to remove unnecessary dimensions
        resume_input = {key: val.squeeze(0) for key, val in resume_input.items()}
        job_input = {key: val.squeeze(0) for key, val in job_input.items()}

        return resume_input, job_input, torch.tensor(label, dtype=torch.float)

# Sample synthetic data for training
resumes = ["Experienced data scientist with expertise in Python and ML.", "Web developer skilled in React and Node.js."]
jobs = ["Looking for a data scientist proficient in Python.", "Hiring a web developer with React skills."]
labels = [1, 0]  # 1 for relevant, 0 for non-relevant

# Prepare dataset and dataloader
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
dataset = ResumeJobDataset(resumes, jobs, labels, tokenizer)
dataloader = DataLoader(dataset, batch_size=2, shuffle=True)

# Initialize model, optimizer, and loss function
model = DualEncoderWithCrossAttention()
optimizer = optim.Adam(model.parameters(), lr=1e-5)

# Training loop
for epoch in range(5):
    for resume_input, job_input, label in dataloader:
        optimizer.zero_grad()
        output = model(resume_input, job_input).squeeze()  # Forward pass
        loss = F.binary_cross_entropy_with_logits(output, label)  # Contrastive loss
        loss.backward()
        optimizer.step()
        print(f"Epoch {epoch+1}, Loss: {loss.item()}")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Function to calculate similarity score between resume and job description
def calculate_similarity(model, resume_text, job_descriptions, tokenizer):
    resume_input = tokenizer(resume_text, padding='max_length', max_length=128, truncation=True, return_tensors="pt")
    similarities = []
    for job in job_descriptions:
        job_input = tokenizer(job, padding='max_length', max_length=128, truncation=True, return_tensors="pt")
        with torch.no_grad():
            similarity_score = model(resume_input, job_input).item()
        similarities.append((job, similarity_score))

    # Sort job descriptions by similarity score
    sorted_jobs = sorted(similarities, key=lambda x: x[1], reverse=True)
    return sorted_jobs[:3]

# Sample job descriptions to compare
job_descriptions = [
    "Looking for a Data Scientist with expertise in Python, Machine Learning, and SQL.",
    "Seeking a Web Developer proficient in React, Django, and Git.",
    "Hiring for a Project Manager with skills in Project Management, Docker, and Git."
]

# Calculate and print the top 3 job matches
top_jobs = calculate_similarity(model, "Project Manager with skills in Project Management, Docker, and Git.", job_descriptions, tokenizer)
print("\nTop 3 job descriptions with the highest similarity:")
for idx, (job, score) in enumerate(top_jobs, start=1):
    print(f"{idx}. Job Description: {job}\n   Similarity Score: {score:.4f}")
```
